```python
df_can = pd.read_excel('./data/Canada.xlsx',
                        sheet_name="Canada by Citizenship",
                        skiprows=(20),
                        skipfooter=2)
print(df_can.head(3))
print(df_can.tail(3))
```

```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
df_can.head(2)
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.head(2)
```

```python
df_can['Total'] = df_can.sum(axis=1)
df_can.head(2)
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.__version__
```

```python
print(plt.style.available) #style plot di list ini https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html

mpl.style.use(['ggplot'])
```

```python
years = range(1980, 2013)
years
```

```python
df_ind_sg = df_can.loc[['Indonesia','Singapore'], years]
df_ind_sg.transpose().plot(kind='line')

plt.title('Indonesia and Singapore immigration to Canada 1980-2013')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
```

```python
df_haiti = df_can.loc['Haiti', years]
df_haiti.plot(kind='line')

plt.title('Haiti immigration to Canada 1980-2013')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')

plt.text(2000,6000, '2010 Earthquake')
```

```python
df_China_India = df_can.loc[['China','India'],years]
df_China_India.transpose().plot()

plt.title('Imigrasi China dan India ke Canada 1980-2013')
plt.ylabel('Jumlah Imigran')
plt.xlabel('Tahun')

plt.show()
```

```python
df_can_sorted = df_can.sort_values(by='Total', axis=0, ascending=False)
df_top5 = df_can_sorted.head(5)
df_top5
```

```python
df_top5.shape
```

```python
df_top5 = df_top5[years].transpose()
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='line', figsize=(28,16))

plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
plt.title('Top 5 Countries Immigrating to Canada 1980-2013')
```

```python
df_top5.head()
```

```python
df_top5.plot(kind='area', figsize=(14,8), stacked=False, alpha=0.2)

plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
plt.title('Top 5 Countries Immigrating to Canada 1980-2013')
```

```python
area_plot_top5 = df_top5.plot(kind='area', figsize=(14,8), stacked=False, alpha=0.2)

area_plot_top5.set_title('Imigrasi top 5 ke Canada 1980-2013')
area_plot_top5.set_ylabel('Jumlah Imigran')
area_plot_top5.set_xlabel('Tahun')
```

```python
df_dns = df_can_sorted.loc[['Denmark','Norway','Sweden'], years].transpose()

count, bin_edges = np.histogram(df_dns, bins=20)

df_dns.plot(kind="hist", bins=20, xticks=bin_edges, figsize=(16,8), alpha=0.3, color=["red","blue","green"])

plt.title("Denmark, Norway, Sweden immigration to Canada 1980-2013")
plt.ylabel("Number of Years")
plt.xlabel("Number of Immigrants")

plt.show()
```

```python
df_gab = df_can_sorted.loc[['Greece','Albania','Bulgaria'],years].transpose()
# df_gab

count, bin_edges = np.histogram(df_gab, 15)

df_gab.plot(kind="hist", bins=15, xticks=bin_edges, alpha=0.35, figsize=(10,8))

plt.title('Histogram of Immigration from Greece, Albania, and Bulgaria from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
plt.show()
```

```python
df_gab.plot(kind="bar", figsize=(20,8))
```

```python
df_iceland = df_can_sorted.loc['Iceland', years]
df_iceland.plot(kind="bar", figsize=(10,8))

plt.title("Iceland Immigration 1980-2013")
plt.xlabel("Years")
plt.ylabel("Number of Immigrants")
plt.show()
```

```python
# top 15 country, horizontal bar plot
df_top15 = df_can_sorted.head(15)['Total']
df_top15.plot(kind="barh", figsize=(10,10))

plt.title("Top 15 countries immigrating to Canada between 1980-2013")
plt.xlabel("Number of Immigrants")
plt.ylabel("Countries")
plt.show()
```

```python
df_continents = df_can_sorted.groupby('Continent', axis=0).sum()
df_continents
```

```python
df_continents['Total'].plot(kind="pie", 
                            figsize=(12,6), 
                            autopct='%1.1f%%', 
                            pctdistance=1.12,
                            labels=None,
                            startangle=90, 
                            explode=[0.1,0,0,0.1,0.1,0.1],
                            colors=['maroon','yellowgreen','lightskyblue','navy','orange','pink'])

plt.axis('equal')
plt.legend(df_continents.index, loc='upper left')
plt.title('Immigration to Canada by Continent [1980 - 2013]', y=1.12) 
plt.show()
```

```python
df_japan_id = df_can_sorted.loc[['Japan', 'Indonesia'], years].transpose()
df_japan_id.plot(kind="box")

plt.title('Box plot of Japanese Immigrants from 1980 - 2013')
plt.ylabel('Number of Immigrants')
plt.xlabel('Countries')

plt.show()
```

```python
df_japan_id = df_can_sorted.loc[['Japan', 'Indonesia'], years].transpose()
df_japan_id.plot(kind="box", vert=False, color='blue')

plt.title('Box plot of Japanese Immigrants from 1980 - 2013')
plt.ylabel('Number of Immigrants')
plt.xlabel('Countries')

plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

df_japan_id.plot(kind="box", ax=ax1)
df_japan_id.plot(kind="box", vert=False, color='blue', ax=ax2)
df_continents['Total'].plot(kind="pie", 
                            figsize=(12,6), 
                            autopct='%1.1f%%', 
                            pctdistance=1.12,
                            labels=None,
                            startangle=90, 
                            explode=[0.1,0,0,0.1,0.1,0.1],
                            colors=['maroon','yellowgreen','lightskyblue','navy','orange','pink'],
                            ax=ax3)
```

```python
# per tahun, total semua negara
df_total = df_can[years].sum(axis=0)
df_total = pd.DataFrame(df_total)
df_total.reset_index(inplace=True)
df_total.columns = ['Year','Total']

df_total.plot(kind="scatter", x='Year', y='Total', figsize=(10,8))

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.ylabel('Number of Immigrants')

plt.show()
```
